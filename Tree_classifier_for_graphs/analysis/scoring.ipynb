```python
import pandas as pd 
from sklearn.metrics import roc_auc_score
import numpy as np
import matplotlib.pyplot as plt


def get_min_score(df):
    min_score = df[df['true_label'] == 1]['predicted_score'].min()
    return (df[df['predicted_score'] < min_score].shape[0] / df[df['true_label'] == 0].shape[0])*100
```
