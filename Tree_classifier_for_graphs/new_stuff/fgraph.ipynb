```python
import pandas as pd


dfs = []

for i in [5,6,7,8,9]:
    df = pd.read_csv(f"/Users/rezadoobary/Documents/MLCorrelator2/ML-correlator/Tree_classifier_for_graphs/new_stuff/features/fgraphs/{i}loops.csv")
    df['loop'] = i
    dfs.append(df)
```

```python
dfs = pd.concat(dfs)
```

```python
dfs['Connectivity_is_connected_DEN'] = dfs['Connectivity_is_connected_DEN'].astype(float)
dfs['Connectivity_is_connected_NUM'] = dfs['Connectivity_is_connected_NUM'].astype(float)
dfs['Connectivity_is_connected_TOTAL'] = dfs['Connectivity_is_connected_TOTAL'].astype(float)
```

```python
import numpy as np
```

```python
dfs = dfs.replace(np.inf, np.nan)
```

```python
feature_cols = [x for x in dfs.columns if 'COEFFICIENTS' != x]
target_cols = 'COEFFICIENTS'
```

```python
X_train = dfs[dfs['loop'].map(lambda x: x in [5, 6, 7,8])]
X_test = dfs[dfs['loop'].map(lambda x: x in [9])]

from xgboost import XGBRFRegressor


model = XGBRFRegressor()
model.fit(X_train[feature_cols].values, X_train[target_cols].map(eval).values)
res = model.predict(X_test[feature_cols].values)


max(res)
```
